# Logistic Regression

As the first approach, we try basic logistic regression. Here, I will select a number of features using my "expert opinion" and the correlation between them and the response and try to build models with them, using cross-validation and AUC scores to decide which one to ultimately go with.

We see the highest coefficients come in Age (higher = more likely to be a bust), offensive rebounding %, rim_fg% (lower = more likely), stl%, ogbpm, dgbpm, ppp, and surprisingly, lane agility.

To create a logistic regression model, we'll need to remove some features to make it simple. However, before that we can use area expertise for this.

- Remove all but offensive efficiency metric (keep PPP)
- Drop all-in-one offensive metric (OGBPM)
- Drop Usage, AST%, and TO% (all future NBA draftees should have the ball in their hands, ast to tov tells playmaking better)
- Drop FT% (maybe can predict shot but not skill)
- Drop 2P%, Mid%, and DUNK% (Rim and 3 are better signals of "touch" and "shooting")
- Drop proportions (useful for classifying, not predicting)
- Drop assist to FG ratio (useful for classifying, not predicting)
- Drop Weight (irrelevant)
- Drop Standing Vertical (one account for height)
- Drop sprint (straight ahead speed doesn't matter)

Then, we'll need to engineer a couple features.
- WINGSPAN DIFFERENTIAL (instead of HEIGHT OR WINGSPAN) to account for positional size
- Turn ncaa_conf into an ordinal ranking of top conferences, mid conferences, and bottom conferences to account for skill level played against.

```python
X_train.loc[X_train["ncaa_conf"].isin(["SEC", "B10", "BE", "ACC", "P12", "P10", "B12"]), "ncaa_conf"] = 3
X_train.loc[X_train["ncaa_conf"].isin(["WCC", "Amer", "A10", "MWC"]), "ncaa_conf"] = 2
X_train.loc[X_train["ncaa_conf"].isin(["CAA", "SB", "MVC", "CUSA", "MAC", "SC", "WAC"]), "ncaa_conf"] = 1
X_train.loc[X_train["ncaa_conf"].isin(["Sum", "MEAC", "OVC", "BSky", "Horz", "ASun", "MAAC", "BW", "NEC", "Pat", "Ivy", "BSth"]), "ncaa_conf"] = 0
```

```python
X_train["WINGSPAN DIFFERENTIAL"] = X_train["WINGSPAN"] - X_train["HEIGHT"]

X_train_1 = X_train[["rookie_Age", "consensus_tier", "ncaa_conf", "ncaa_oreb%", "ncaa_dreb%", 
                     "ncaa_3p%", "ncaa_blk%", "ncaa_stl%", "ncaa_ftr", "ncaa_foul_rate", 
                     "ncaa_recruit_rank", "ncaa_ast_tov_ratio", "ncaa_rim_fg%", 
                     "ncaa_dgbpm", "ncaa_ppp", "WINGSPAN DIFFERENTIAL", "VERTICAL LEAP", 
                     "LANE AGILITY"
                    ]]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_1 = scaler.fit_transform(X_train_1)
```

Last, before we start working with the data, we need to define the training set and work with that to find the most generalized and accurate model before moving to the test set.

```python
def cv_log_score(X, y, c=10, scoring="f1"):
    from sklearn.linear_model import LogisticRegression
    from sklearn.model_selection import cross_val_score
    
    model = LogisticRegression(penalty="none", random_state=9, max_iter=1000)
    scores = cross_val_score(model, X, y, scoring=scoring, cv=c)
    return np.mean(scores)
```

We know interaction terms will be useful, as we know younger players or longer players tend to be desired by teams. However youth or length on its own isn't necessarily indicative of success. However, youth that can do a certain thing at a high level, or a long player that can do another thing could be a more successful bet.

- AGE and PPP (marks scoring efficiency at a certain age)
- AGE and A2T (tries to measure passing ability at a young age)
- AGE and FT% (we know FT% predicts NBA shooting success, so tries to ID shooting at young age)
- AGE and recruit rank (as good recruits stay in school longer, their former ranking matters less)
- WINGSPAN DIFF and DGBPM (identifies a lengthy defender [plus in NBA])
- WINGSPAN DIFF and FT% (identifies a lengthy player that can shoot)

# Regularized Logistic Regression

As we couldn't make meaningful improvements using domain knowledge for feature selection, we'll use regularized logistic regression to attempt to wrangle useful and rogue variables in a single model. 

As we are now going to work with a different kind of logistic regression, we need to create a new assessment function.

```python
def cv_log_score(X, y, cn=10, scoring="f1", reg_type="l1", c=1, solver="liblinear"):
    from sklearn.linear_model import LogisticRegression
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import RepeatedStratifiedKFold
    
    cv = RepeatedStratifiedKFold(n_splits=cn, n_repeats=5, random_state=9)
    model = LogisticRegression(penalty=reg_type, C=c, random_state=9, max_iter=1000, solver=solver)
    scores = cross_val_score(model, X, y, scoring=scoring, cv=cv)
    
    return np.mean(scores)
```

```python
X_train
```

```python
X_train_l1 = X_train.drop(["GMM_Primary_Cluster", "GMM_Secondary_Cluster", "Defensive Specialist", "Floor-Spacing Wing", "Lead Guard", "Pass-First Playmaker", "Rim Runner", "Sharpshooting Guard", "Skilled Forward"], axis=1)
```

```python
for i in [.001, .005, 0.01, .05, .1, .5, 1, 5]:
    print(f"Accuracy Score for C={i}: {cv_log_score(X_train_l1, y_train, c=i, scoring='accuracy')}")
```

```python
for i in [0.01, .05, .1, .5, 1]:
    print(f"F1 Score for C={i}: {cv_log_score(X_train_l1, y_train, c=i)}")
```

After trying different values for C, we settle on .1, as it provides a high regularization and F1 score.

This is a meaningful improvement over the first model, but now as we're done with regularized models, we'll switch to tree models.

# Random Forest

It's tough to implement the text-based features into our model, so we'll drop the clusters here.

Then, we'll procede with a random forest model to predict.

```python
X_train = X_train.drop(["GMM_Primary_Cluster", "GMM_Secondary_Cluster"], axis=1)
```

```python
X_train = X_train.iloc[:, :-7]
```

```python
X_train
```

To tune the model, we'll follow the steps of choosing hyper parameters from the guide on "machinelearningmastery.com"

That means we will first tune the number of samples.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
    
for i in [60, 70, 80, 90, 100]:
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=9)
    model = RandomForestClassifier(max_samples=i, random_state=9)
    scores = cross_val_score(model, X_train, y_train, scoring="f1", cv=cv)
    print(f"RF F1 Score w/ {i}% of samples: {np.mean(scores)}")
```

We'll use a max samples value of 60%, as it yielded strongest results.

Next, we need to explore the number of trees (n_estimators).

```python
for i in [10, 50, 100, 500]:
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=9)
    model = RandomForestClassifier(n_estimators=i, max_samples=60, random_state=9)
    scores = cross_val_score(model, X_train, y_train, scoring="f1", cv=cv)
    print(f"RF F1 Score w/ {i} estimators: {np.mean(scores)}")
```

N estimators performs best with 500 "trees" in the forest. This becomes an expensive model so we'll stick with 100.

Next, we need to identify the "max_depth" of our trees.

```python
for i in [1, 2, 3, 4, 5, 6, 7, None]:
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=9)
    model = RandomForestClassifier(n_estimators=100, max_depth=i, max_samples=60, random_state=9)
    scores = cross_val_score(model, X_train, y_train, scoring="f1", cv=cv)
    print(f"RF F1 Score w/ {i} depth: {np.mean(scores)}")
```

We find that a max tree depth of 4 yields best results marginally.

Last, we need to know the number of features that creates the strongest model. With 37 features, the current value is set at about 6. However, this could be better if lower or higher, so we'll try values between 4 and 8.

```python
for i in [4, 5, 6, 7, 8]:
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=9)
    model = RandomForestClassifier(max_depth=4, max_samples=60, max_features=i, random_state=9)
    scores = cross_val_score(model, X_train, y_train, scoring="f1", cv=cv)
    print(f"RF F1 Score w/ {i} features: {np.mean(scores)}")
```

6 gives us the best F1 score and keeps the model the same as the 6 it would've had. With our final values, let's try the n_estimators one more time and see if we can fit a similarly performing cheaper model.

Ultimately, the F1 score is still lower than the regularized logistic model, meaning we'll stick with the regularized one.
