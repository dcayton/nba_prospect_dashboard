```python
import numpy as np
import pandas as pd

import os
import sys

current = os.path.dirname(os.path.realpath("peak_standard_scaler"))
parent = os.path.dirname(current)

pd.set_option("display.max_columns", 150)
peaks = pd.read_csv(parent + "/player_data/peaks.csv")
peaks.head()
```

```python
peaks_scaler_stats = ["peak_AST_2_TOV", "peak_AST_2_FG", "peak_AST%", "peak_TOV%", "peak_USG%", "peak_PPP", "peak_2P_prop",
                      "peak_0-3_prop", "peak_3-10_prop", "peak_10-16_prop", "peak_16-3P_prop", "peak_3P_prop",
                      "peak_corner_%3PA", "peak_2P_%astd", "peak_3P_%astd", "peak_0-3_fg%",
                      "peak_3-10_fg%", "peak_10-16_fg%", "peak_16-3P_fg%", "peak_3P_fg%", "peak_corner_3P%",
                      "peak_STL%", "peak_BLK%", "peak_ORB%", "peak_DRB%"]
peaks_scale = peaks.loc[:,peaks_scaler_stats]
```

```python
peaks_scale["peak_3P_%astd"] = peaks_scale["peak_3P_prop"] * peaks_scale["peak_3P_%astd"]
peaks_scale["peak_2P_%astd"] = peaks_scale["peak_2P_prop"] * peaks_scale["peak_2P_%astd"]

peaks_scale["peak_corner_%3PA"] = peaks_scale["peak_corner_%3PA"] * peaks_scale["peak_3P_prop"]
peaks_scale["peak_3P_prop_not_corner"] = peaks_scale["peak_3P_prop"] - peaks_scale["peak_corner_%3PA"]

peaks_scale = peaks_scale.drop(["peak_3P_prop", "peak_2P_prop"], axis=1)
```
